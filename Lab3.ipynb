```python
import keras
from keras.datasets import mnist
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Dense, Flatten
from keras.models import Sequential
from sklearn.pipeline import Pipeline
from keras import layers, models
from keras.optimizers import Adam
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

**1.2 Loading the Training adn Out-of-sample Data**

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

# **2. Data Preprocessing**

2.1 One Hot Encoding for categorical data

```python
# No. of classes = max(y) + 1
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
```

# **3. Defining the CNN Model**

**3.1 LeNet Model Architecture**

```python
def build_model(filters, learning_rate):
  model = models.Sequential()
  #define filters and convolutional layers here
  model.add(layers.Conv2D(filters=filters, kernel_size=(3, 3), activation='relu',
  input_shape=(28, 28, 1)))
  #Add a maxpooling layer
  model.add(layers.MaxPooling2D(pool_size=(2, 2)))
  #Flatten the output and give it to a fully connected layer
  model.add(layers.Flatten())
  #one hidden layer maps the flattened neurons to output
  model.add(layers.Dense(10, activation='softmax'))
  optimizer = Adam(learning_rate=learning_rate)
  model.compile(optimizer=optimizer, loss='categorical_crossentropy',
  metrics=['accuracy'])
  return model
```

# **4. Parameter Exploration**

**4.1 Using GridSearchCV for stratified cross validation**

```python
# Using Keras Classifier API to make our Keras model compatible with GridSearchCV
model = KerasClassifier(build_fn=build_model,
                        filters=16,
                        learning_rate=0.001,
                        verbose=1)
pipeline = Pipeline(steps=[('model', model)])
learning_rate=[0.001, 0.01]
filters = [16, 32]
param_grid = dict(filters=filters, learning_rate=learning_rate)
grid = GridSearchCV(estimator=pipeline.named_steps['model'], param_grid=param_grid, cv=5)
grid_result = grid.fit(x_train, y_train)
```

**4.2 Best accuracy and parameters**

```python
[best_score, best_params] = grid_result.best_score_, grid_result.best_params_
print([best_score, best_params])
```

When our CNN model has 32 filters and 0.001 learning rate, our model's training accuracy is 96.23%

# **5. Training the best model**

Training our CNN model using the best parameters on out-of-sample data

```python
best_model=KerasClassifier(build_fn=build_model,
                        filters=32,
                        learning_rate=0.001,
                        verbose=1)
pipeline = Pipeline(steps=[('model', best_model)])
pipeline.fit(x_train, y_train)
```

# **6. Evaluating the best model's performance**

Using out-of-sample data set to for making predictions and evaluating the best model's performance on it.
- x_test : out-of-sample values used for prediction
- y_predicted : predictions made on x_test
- accuracy_score : accuracy of the best model is measured based on how accurate y_predicted is, given y_test are the true values

```python
y_predicted = pipeline.predict(x_test)
accuracy = accuracy_score(y_test, y_predicted)
```

```python
print(accuracy)
```

The validation accuracy (accuracy on out-of-sample data) of our model using the best parameters (32 filters and 0.001 learning rate) is 96.85%
